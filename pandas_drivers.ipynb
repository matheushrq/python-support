```python
import pandas as pd

drivers = pd.read_csv('drivers.csv', usecols = [0, 4, 5])
drivers
```

```python
# exibindo somente os pilotos que estiveram na temporada de 2020
drivers [drivers.year == 2020] # acessando a coluna como propriedade do dataframe
```

```python
# selecionando os pilotos que foram da Ferrari entre 2004 e 2009 passando as colunas como variáveis
drivers[(drivers['team'] == 'Ferrari') &
        (drivers['year'] > 2004) &
        (drivers['year'] < 2009)][['driver', 'team', 'year']]
```

```python
# selecionando os pilotos que foram da Ferrari entre 2004 e 2009 passando as colunas como variáveis ordenado pelo ano
drivers[(drivers['team'] == 'Ferrari') &
        (drivers['year'] > 2004) &
        (drivers['year'] < 2009)][['driver', 'team', 'year']].sort_values(by = 'year')
```

```python
# selecionando os pilotos que foram da Ferrari entre 2004 e 2009 pelo método query
drivers.query('team == "Ferrari" and year > 2004 and year < 2009')[['driver', 'team', 'year']]
```

```python
# selecionando os pilotos que foram da Ferrari entre 2004 e 2009 pelo método query ordenado pelo ano
drivers.query('team == "Ferrari" and year > 2004 and year < 2009')[['driver', 'team', 'year']].sort_values(by = 'year')
```
